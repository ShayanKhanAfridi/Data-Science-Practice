# **Advanced Matplotlib**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# **Colored Scatter Plots**

**Step#01:** replace the categories with numbers (0,1,2,3,..)

**Step#02:** pass the categories column as colors e.g c=iris['Species']

**Step#03:** plt.colorbar()

**Step#04:** cmap='jet'/'summer'/'winter'/etc etc

```python
iris = pd.read_csv("iris.csv")
iris.sample(5)
```

```python
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'])

plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
```

```python
iris['Species'] = iris['Species'].replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
iris.sample(5)
```

```python
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c=iris['Species'], cmap='jet', alpha=0.8)  #alpja means opacity

plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
```

### **plot size**

```python
#plot size
plt.figure(figsize=(10,5))
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c=iris['Species'], cmap='jet', alpha=0.8)  #alpja means opacity

plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
```

### **Annotations**

```python
batters = pd.read_csv('batter.csv')
batters.head()
```

```python
sample_df = batters.head(100).sample(25, random_state=5)
```

```python
sample_df
```

```python
plt.figure(figsize=(14,9))
plt.scatter(sample_df['avg'], sample_df['strike_rate'], s=sample_df['runs'])

for i in range(sample_df.shape[0]):
  plt.text(sample_df['avg'].values[i], sample_df['strike_rate'].values[i], sample_df['batter'].values[i])
```

### **Horizontal and Vertical Lines**


*   plt.axhline(130, color='red')
*   plt.axvline(30, color='red')





```python
plt.figure(figsize=(14,9))
plt.scatter(sample_df['avg'], sample_df['strike_rate'], s=sample_df['runs'])

plt.axhline(130, color='red')
plt.axhline(140, color='green')
plt.axvline(30, color='red')

for i in range(sample_df.shape[0]):
  plt.text(sample_df['avg'].values[i], sample_df['strike_rate'].values[i], sample_df['batter'].values[i])
```

# **Subplots**

```python
# A diff way to plot graphs
batters.head()
```

```python
#normal way
plt.figure(figsize=(10,5))

plt.scatter(batters['avg'], batters['strike_rate'])

plt.title("Avg vs Strike rate")
plt.xlabel("Average")
plt.ylabel("Strike rate")

plt.show()
```

### **Plotting graphs using [fig, ax = plt.subplots()]**

```python
#another way of plotting graphs
fig, ax = plt.subplots(figsize=(10,5))

ax.scatter(batters['avg'], batters['strike_rate'], color='red')

ax.set_title("Avg vs Strike rate")
ax.set_xlabel("Average")
ax.set_ylabel("Strike rate")

fig.show()
```

### **plotting 2 graphs side by side**

```python
#plotting 2 graphs side by side
fig, ax = plt.subplots(nrows=2, ncols=1, sharex=True, figsize=(10,5))

ax[0].scatter(batters['avg'], batters['strike_rate'], color='blue')
ax[1].scatter(batters['avg'], batters['runs'], color='red')

ax[0].set_title("Avg vs Strike rate")
ax[0].set_ylabel("Strike rate")


ax[1].set_title("Avg vs Runs")
ax[1].set_xlabel("Average")
ax[1].set_ylabel("Runs")

fig.show()
```

### **plotting 4 graphs**

```python
#plotting 4 graphs
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(12,8))

#Scatter plots
ax[0,0].scatter(batters['avg'], batters['strike_rate'], color='blue')
ax[0,1].scatter(batters['avg'], batters['runs'], color='red')

#Histograms
ax[1,0].hist(batters['avg'], edgecolor='black')
ax[1,1].hist(batters['strike_rate'], edgecolor='black')

fig.show()
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(2,2,1)
ax1.scatter(batters['avg'],batters['strike_rate'],color='red')

ax2 = fig.add_subplot(2,2,2)
ax2.hist(batters['runs'], edgecolor='black')

ax3 = fig.add_subplot(2,2,3)
ax3.hist(batters['avg'], edgecolor='black')

fig.show()
```

# **3D Scatter Plots**

```python
fig = plt.figure(figsize=(12,7))

ax = plt.subplot(projection='3d')

ax.scatter3D( batters['runs'], batters['avg'], batters['strike_rate'])

ax.set_xlabel('Runs')
ax.set_ylabel('Average')
ax.set_zlabel('Strike rate')
```

# **3D Line Plot**

```python
x = [0,12,15,24]
y = [0,10,15,20]
z = [0,8,14,19]

fig = plt.figure()

ax = plt.subplot(projection='3d')

ax.scatter3D(x,y,z, s=[100,100,100,100])
ax.plot3D(x,y,z, color='red')
```
