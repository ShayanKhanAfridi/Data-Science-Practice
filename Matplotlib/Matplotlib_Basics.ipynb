# **Bar Chart**
- Bivariate Analysis
- Numerical vs Categorical
- Use case - Aggregate analysis of groups

### **simple bar chart**

```python
# simple bar chart
children = [10,20,40,10,20]
colors = ['red','blue','green','yellow','orange']

plt.bar(colors,children,color = colors)
plt.title('Colors Like by childrens')
plt.xlabel('Colors')
plt.ylabel('No. of Childrens')
```

### **horizontal bar chart**

```python
#horizontal bar chart
plt.barh(colors,children,color = colors)
plt.title('Colors Like by childrens')
plt.xlabel('Colors')
plt.ylabel('No. of Childrens')
```

### **Multiple bar charts**

```python
#multiple bar charts
df = pd.read_csv('/content/batsman_season_record.csv')
df
```

```python
#When multiple bar plots are drawn using the same x-values, the bars overlap and the chart becomes unclear.
plt.bar(df['batsman'], df['2015'])
plt.bar(df['batsman'], df['2016'])
plt.bar(df['batsman'], df['2017'])
```

```python
#To fix this, we shift each bar slightly along the x-axis using an offset (like x - width, x, x + width) to display multiple bars side by side.
plt.bar(np.arange(df.shape[0]) - 0.2, df['2015'], width = 0.2)
plt.bar(np.arange(df.shape[0]), df['2016'], width = 0.2)
plt.bar(np.arange(df.shape[0]) + 0.2, df['2017'], width = 0.2)

plt.xticks(np.arange(df.shape[0]), df['batsman'])  #replacing the bar labels

plt.title('Batsman Season Record')
plt.xlabel('Batsman', color = 'blue')
plt.ylabel('Runs', color = 'blue')
plt.show()
```

### **xticks**

```python
# A problem (ziada bara naam hoga labels me tu wo overlap hojaega)
children = [10,20,40,10,20]
colors = ['red red red red red','blue blue blue blue blue','green green green green green','yellow yellow yellow yellow yellow','orange orange orange orange orange']

plt.bar(colors,children)

#Solution
plt.xticks(rotation='vertical')
plt.show()
```

### **Stacked Bar Chart**

```python
plt.bar(df['batsman'], df['2015'], label = '2015')
plt.bar(df['batsman'], df['2016'], bottom=df['2015'], label = '2016')
plt.bar(df['batsman'], df['2017'], bottom=(df['2016'] + df['2015']), label = '2017')

plt.legend()
plt.title('Batsman Season Record')
plt.xlabel('Batsman')
plt.ylabel('Runs')
plt.show()
```

### **logarithmic scale**
A logarithmic scale in a histogram is used to display data with a large range of values more clearly.
It is applied using log=True in plt.hist(), which changes the y-axis to a logarithmic scale.

```python
# logarithmic scale
arr = np.load('/content/big-array.npy')
arr.shape

plt.hist(arr,bins=[10,20,30,40,50,60,70], log=True, edgecolor='black')
plt.show()
```

# **Pie Chart**
- Univariate/Bivariate Analysis
- Categorical vs numerical
- Use case - To find contibution on a standard scale

### **simple data**

```python
#simple data
data = [23,45,100,20,49]
subjects = ['english','maths','science','sst','computer']

plt.pie(data, labels = subjects)
plt.show()
```

### **on data**

```python
# on data
df = pd.read_csv('/content/gayle-175.csv')
df

plt.pie(df['batsman_runs'], labels = df['batsman'], autopct='%0.1f%%')

plt.show()
```

### **percentage and colors**

```python
#percentage and colors
plt.pie(df['batsman_runs'], labels = df['batsman'], autopct='%0.1f%%', colors=['blue','green','yellow','pink','cyan','brown'])

plt.show()
```

### **explode and shadow**

```python
#explode and shadow
plt.pie(df['batsman_runs'], labels = df['batsman'], autopct='%0.1f%%', explode=[0.1,0,0,0,0,0], shadow=True)

plt.show()
```
