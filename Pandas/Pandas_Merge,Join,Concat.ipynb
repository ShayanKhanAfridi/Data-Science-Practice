````python
import numpy as np
import pandas as pd
````

# **Pandas Merging, Joining and Concatenating**

````python
courses = pd.read_csv('/content/courses.csv')
students = pd.read_csv('/content/students.csv')
nov = pd.read_csv('/content/reg-month1.csv')
dec = pd.read_csv('/content/reg-month2.csv')
matches = pd.read_csv('/content/matches.csv')
delivery = pd.read_csv('/content/deliveries.csv')
````

## **Concat**

### **#pd.concat**

````python
#pd.concat
reg = pd.concat([nov,dec],ignore_index=True) #nov aur december ka data vertically merge hogaya
reg
````

### **#mullitindex -> fetch using loc**

````python
# mullitindex DataFrame -> fetch using loc
multi = pd.concat([nov,dec], keys = ["Nov","Dec"])
multi.loc["Nov"]
multi.loc["Dec"]
multi.loc[("Nov",1)]
````

### **#Concat Horizontally (axis = 1)**

````python
#Concat Horizontally (axis = 1)
pd.concat([nov,dec],axis = 1)
````

df1.merge(df2, how = "inner/outer/left/right", on = "unique column")

### **#inner join**

````python
#inner join
students.merge(reg, how = "inner", on="student_id")
````

### **#left join**

````python
#left join
courses.merge(reg, how = "left", on = "course_id")
````

### **#right join**

````python
temp_df = pd.DataFrame({
    'student_id':[26,27,28],
    'name':['Nitish','Ankit','Rahul'],
    'partner':[28,26,17]
})

students = pd.concat([students,temp_df],ignore_index=True)
````

````python
#right join
students.merge(reg, how = "right", on = "student_id")
````

### **#outer join**

````python
#outer join
students.merge(reg, how="outer", on="student_id")
````

## **Practice Questions**

### **1. find total revenue generated**

````python
# 1. find total revenue generated
total = courses.merge(reg, how = "inner", on = "course_id")["price"].sum()
total
````

### **2. find month by month revenue**

````python
# 2. find month by month revenue
temp_df = pd.concat([nov,dec], keys = ["Nov", "Dec"]).reset_index()
temp_df.head()
````

````python
temp_df.merge(courses, on = "course_id").groupby("level_0")["price"].sum()
````

### **3. Print the registration table**

````python
# 3. Print the registration table
# cols -> name -> course -> price
reg.merge(students,on = "student_id").merge(courses,on="course_id")[["name","course_name","price"]]
````

### **4. Plot bar chart for revenue/course**

````python
# 4. Plot bar chart for revenue/course
reg.merge(courses, how="inner", on="course_id").groupby("course_name")["price"].sum().plot(kind = "bar")
````

### **5. find students who enrolled in both the months**

````python
# 5. find students who enrolled in both the months
common_st_id = np.intersect1d(nov["student_id"], dec["student_id"])
common_st_id
````

````python
students[students["student_id"].isin(common_st_id)]
````

### **6. find course that got no enrollment**

````python
# 6. find course that got no enrollment
# courses['course_id']
# regs['course_id']
course_id_list = np.setdiff1d(courses["course_id"],reg["course_id"])
courses[courses["course_id"].isin(course_id_list)]
````

### **7. find students who did not enroll into any courses**

````python
# 7. find students who did not enroll into any courses
student_id_list = np.setdiff1d(students['student_id'],reg['student_id'])
students[students['student_id'].isin(student_id_list)]
````

### **8. Print student name -> partner name for all enrolled students**

````python
# 8. Print student name -> partner name for all enrolled students
#self join
students.merge(students, how="inner", left_on="partner", right_on="student_id")[["name_x","name_y"]]
````

### **9. find top 3 students who did most number enrollments**

````python
# 9. find top 3 students who did most number enrollments
reg.merge(students, on="student_id").groupby(["student_id","name"])["name"].count().sort_values(ascending=False).head(3)
````

### **10. find top 3 students who spent most amount of money on courses**

````python
# 10. find top 3 students who spent most amount of money on courses
reg.merge(courses, on="course_id").merge(students, on="student_id").groupby(["student_id","name"])["price"].sum().sort_values(ascending=False).head(3)
````

### **find top 3 studiums with highest sixes/match ratio**

````python
# find top 3 studiums with highest sixes/match ratio
matches.head(2)
````

````python
delivery.head(2)
````

````python
#merge kardiye 2no table
temp_df = delivery.merge(matches, left_on="match_id", right_on="id")

#unhi balls ka data jinpar chakka para
six_df = temp_df[temp_df["batsman_runs"]==6]

#har stadium me kitne chakke pare hain
num_sixes = six_df.groupby("venue")["venue"].count()

#har stadium me kitne matches howe hain
num_matches = matches["venue"].value_counts()
````

````python
(num_sixes/num_matches).sort_values(ascending=False).head(3)
````

### **find orange cap holder of all the seasons**

````python
# find orange cap holder of all the seasons
temp_df = delivery.merge(matches, left_on="match_id", right_on="id")
temp_df.groupby(["season","batsman"])["batsman_runs"].sum().reset_index().sort_values('batsman_runs', ascending=False).drop_duplicates(subset=['season'],keep='first').sort_values('season')
````
