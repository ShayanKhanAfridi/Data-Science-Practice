# **Vectorized String**

### **What are Vectorized Operations**
Vectorized operations are operations that let you perform calculations on an entire array or column at once, instead of processing each element one by one.

```python
# What are vectorized operations
a = np.array([1,2,3,4])
a * 5  #yahan par hamne array par aik single operation chalaya jo array ke har item par chala)
```

### **problem in vectorized opertions in vanilla python**

```python
# problem in vectorized opertions in vanilla python
s = ['cat','mat',None,'rat']

#[i.startswith('c') for i in s]  #error dega cuz list me None bhi hai
```

### **How Pandas Solve this issue**

```python
# How pandas solves this issue?
s = pd.Series(['cat','mat',None,'rat'])

# string accessor
s.str.startswith('c')

# fast and optimized
```

```python
df = pd.read_csv('/content/titanic.csv')
df.head()
```

## **Common String Functions**

```python
# lower/upper/capitalize/title
df['Name'].str.lower()
df['Name'].str.upper()
df['Name'].str.capitalize()
df['Name'].str.title()

#len
df['Name'][df['Name'].str.len() == 82]

#strip
"      Shayan    ".strip()
df['Name'].str.strip()  # saray spaces removes kar dega
```

### **split**

```python
#split -> get
df['last_name'] = df['Name'].str.split(',').str.get(0)  #split se aik list ban gai (containg surname and name) .get(0 kia tu sirf surname ajaega)
```

```python
df.head()
```

```python
df[['Title','first_name']] = df['Name'].str.split(',').str.get(1).str.strip().str.split(' ', n=1, expand = True) #n=1 ka mtlb ke sirf pehle space ko split karega
df.head()
```

```python
df.head()
```

### **replace**

```python
#replace
df['Title'] = df['Title'].str.replace('Ms','Miss')
df['Title'] = df['Title'].str.replace('Mlle','Miss')

df['Title'].value_counts()
```

### **Filtering**

```python
# startswith/endswith
df[df['first_name'].str.startswith('A')]

df[df['first_name'].str.endswith('A')]

#isdigit
df[df['first_name'].str.isdigit()]
```

### **applying regex**

```python
# applying regex
# contains
# search john -> both case
df[df['first_name'].str.contains('john', case=False)].head()
```

```python
# find lastnames with start and end char vowel
df[df['last_name'].str.contains('^[aeiouAEIOU].+[aeiouAEIOU]$')].head()
```
